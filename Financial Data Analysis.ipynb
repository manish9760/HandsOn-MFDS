```python
# To know data types
df.dtypes
```

```python
#for print 5 rows data
df.head()
```

```python
# MEAN
mean=df['Mar Cap - Crore'].mean()
print(f"Mean of Mar Cap-Crore = {mean}")

mean=df['Sales Qtr - Crore'].mean()
print(f"Mean of Sales Qtr-Crore={mean}")
```

```python
# MEDIAN
median=df['Mar Cap - Crore'].median()
print(f"Median of Mar Cap- Crore= {median}")

median=df['Sales Qtr - Crore'].median()
print(f"Median of Sales Qtr-Crore= {median}")
```

```python
# Variance

Variance=df['Mar Cap - Crore'].var()
print(f"Variance of Mar Cap-Crore= {Variance}")

Variance=df['Sales Qtr - Crore'].var()
print(f"Variance of Sales Qtr-Crore= {Variance}")
```

```python
# Standard Deviation

std=df['Mar Cap - Crore'].std()
print(f"Standard Deviation of Mar Cap- Crore= {std}")

std=df['Sales Qtr - Crore'].std()
print(f"Standard Deviation of Sales Qtr-Crore= {std}")
```

```python
# MODE

mode=df['Mar Cap - Crore'].mode()
print(f"Mode of Mar Cap- Crore= {mode}")

mode=df['Sales Qtr - Crore'].mode()
print(f"Mode of Sales Qtr-Crore= {mode}")

```

```python
# Skewness
skewness=df['Mar Cap - Crore'].skew()
print(f"Skewness of Mar cap-Crore= {skewness}")

skewness=df['Sales Qtr - Crore'].skew()
print(f"Skewness of Mar Cap-Crore= {skewness}")
```

```python
# Kurtosis
kurtosis=df['Mar Cap - Crore'].kurt()
print(f"Kurtosis of Mar Cap-Crore= {kurtosis}")

kurtosis=df['Sales Qtr - Crore'].kurt()
print(f"Kurtosis of Sales Qtr-Crore= {kurtosis}")
```

```python
# Range
range= df['Mar Cap - Crore'].max() - df['Mar Cap - Crore'].min()
print(f"Range of Mar Cap- Crore= {range}")

range= df['Sales Qtr - Crore'].max() - df['Sales Qtr - Crore'].min()
print(f"Range of Mar Cap-Crore= {range}")
```

```python
#Graph Representation


df['Mar Cap - Crore'].value_counts().plot(kind='bar',title='Mar Cap - Crore')
plt.xlabel('Count')
plt.ylabel('Mar Cap - Crore')
plt.show()

df['Sales Qtr - Crore'].value_counts().plot(kind='bar',title='Sales Qtr - Crore')
plt.xlabel('Count')
plt.ylabel('Sales Qtr - Crore')
plt.show()
```
